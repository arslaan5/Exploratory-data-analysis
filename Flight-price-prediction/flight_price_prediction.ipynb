```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
train_df = pd.read_excel('train_data.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('test_data.xlsx')
test_df.head()
```

# **Feature Engineering**

```python
final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'])
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date_of_Journey'].dt.day.astype(int)
final_df['Month'] = final_df['Date_of_Journey'].dt.month.astype(int)
final_df['Year'] = final_df['Date_of_Journey'].dt.year.astype(int)
final_df.info()
```

```python
final_df.head()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
final_df.head()
```

```python
# final_df = final_df.drop('Arr_time', axis=1)
```

```python
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_Min'] = final_df['Arrival_Time'].str.split(':').str[1].astype(int)
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Departure_Hour'] = final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Departure_Min'] = final_df['Dep_Time'].str.split(':').str[1].astype(int)
final_df.head()
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```

```python
final_df.head()
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration_Hours'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_Minutes'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
final_df.head()
```

```python
final_df['Duration_Minutes'].isnull().sum()
```

```python
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

```python
final_df[final_df['Duration_Hours']=='5m']
```

```python
final_df['Duration_Minutes'] = final_df['Duration_Minutes'].fillna(0)
```

```python
final_df['Duration_Hours'] = final_df['Duration_Hours'].astype(int)
final_df['Duration_Minutes'] = final_df['Duration_Minutes'].astype(int)
```

```python
final_df.drop('Duration', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration_Hours'] = final_df['Duration_Hours']*60
```

```python
final_df['Duration_Minutes'] = final_df['Duration_Hours'] + final_df['Duration_Minutes']
final_df.head()
```

```python
final_df.drop('Duration_Hours', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Duration_in_Min'] = final_df['Duration_Minutes']
final_df.head()
```

```python
final_df.drop('Duration_Minutes', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
lbe = LabelEncoder()

final_df['Airline'] = lbe.fit_transform(final_df['Airline'])
final_df['Source'] = lbe.fit_transform(final_df['Source'])
final_df['Destination'] = lbe.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = lbe.fit_transform(final_df['Additional_Info'])

final_df.head()
```
